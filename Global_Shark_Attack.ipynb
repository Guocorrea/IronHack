## Criar um DataFrame com os acidentes registrados entre 1990 e 2020.

```python
# Para reinicar tudo que foi feito acima, utilizar essa célula.
import pandas as pd
import numpy as np
# Baixar o Arquivo
atck_xls = pd.read_excel('Datas/GSAF5.xls')

# Criar o DataFrame com as informações que parecem mais úteis
atck = (atck_xls['Year'], atck_xls['Type'], atck_xls['Country'], atck_xls['Area'],
        atck_xls['Activity'],atck_xls['Sex '], atck_xls['Age'], atck_xls['Fatal (Y/N)'], 
        atck_xls['Time'], atck_xls['Species '])
atck = pd.DataFrame(atck)
atck = atck.transpose()

# Transformar a coluna 'Year' (nossa variável chave), de object para float
atck['Year'] = atck['Year'].astype(float, errors = 'raise')

# A primeira linha está classificado como 'No Shark involvement', vou deletar. 
# IMPORTANTE: depois ver se tem mais 'No Shark involvement' na coluna 'Injury'!!!
atck = atck.drop(0, axis = 0)
atck = atck.reset_index(drop = True)

# --------------- AQUI COMEÇAM OS CÓDIGOS PARA CRIAR O DATA FRAME 1990 - 2020 ------------------
# Criar uma máscara para criar o DataFrame com os ataques de 1990 até 2020.
atck90 = atck[atck['Year'] > 1989]
atck = atck90[atck90['Year'] < 2021]

# Organizar a tabela de acordo com o ano em que aconteceu o ataque, em ordem decrescente -
# e pela idade em ordem crescente
# Resetar a indexação das linhas
atck = atck.sort_values(['Year', 'Age'], ascending = [False, True])
atck = atck.reset_index(drop = True)

```

```python
def info_mult(linha):
    '''Criar uma lista com todas as informações do index em que tem mais de uma informação
    no campo AGE'''
    
    infos = list()
    global atck
    infos.append(atck.iloc[linha, 0])
    infos.append(atck.iloc[linha, 1])
    infos.append(atck.iloc[linha, 2])
    infos.append(atck.iloc[linha, 3])
    infos.append(atck.iloc[linha, 4])
    infos.append(atck.iloc[linha, 5])
    infos.append(atck.iloc[linha, 6])
    infos.append(atck.iloc[linha, 7])
    infos.append(atck.iloc[linha, 8])
    infos.append(atck.iloc[linha, 9])    
    
    return infos
```
