## Tirar os elementos da coluna 'Age' que apresentam dúvida de idade. Ex: 18 or 20.
-------------------------------------------------
Vou utilizar a menor idade. Alteração não significativa para a estatística final.

```python
# Encontrar todos os elementos da atck['Age'] com idade 'x or y'
import regex as re

padrão = '[ 0-9]+[ or]+[ 0-9]+'
lista_or = list()
for i in atck['Age']:
    try:
        if i in re.findall(padrão, i):
            lista_or.append(i)
    except:
        None
print(lista_or)
print(len(lista_or))
```

```python
# Criar uma lista com o index de cada linha de casos múltiplos
lista_index = list()
atck_copy = atck[:]

for i in lista_or:
    g = atck_copy[atck_copy['Age'] == i]
    h = g['index']
    if len(h) == 1:
        lista_index.append(int(h.values))
        atck_copy.drop(int(h.values))
    elif len(h) > 1:
        for i in h:
            ab = h[i]
            lista_index.append(ab)
                
# Apareceram na lista valores dobrados. Vou criar um set para corrigir!

set_index = set(lista_index)
print(set_index)
print(len(set_index))
```

```python
# Utilizar o set_index para alterar a idade das células dúbias.
for i in set_index:
    atck['Age'].iloc[i] = atck['Age'].iloc[i][0] + atck['Age'].iloc[i][1]

# Células dúbias da coluna 'Age' resolvidas!
```

## Dividindo informações múltiplas em uma mesma linha
----------------------------------------------------------------------------------
Encontrei algumas linhas com informações multiplas, ou seja, mais de um acidente registrado em uma única informação.



### Começar identificando todas as linhas com informações múltiplas. 
 - Encontrar as linhas com informações múltiplas
 - Registrar a indexação delas
 - Separar a informação múltipla em informações únicas

```python
# Encontrar as linhas que tem 2 vítimas registradas

import regex as re

padrão = '[ 0-9]+[, &]+[ 0-9]+'
lista_mult2 = list()
for i in atck['Age']:
    try:
        if i in re.findall(padrão, i):
            lista_mult2.append(i)
    except:
        None
print(lista_mult2)
print(len(lista_mult2))
```

```python
# Econtrar a indexação com as informações das linhas com informações múltiplas com 2 vítimas
# Vou precisar dessa informação para criar as linhas com informações únicas correspondentes
lista_index2 = list()

for elemento in lista_mult2:
    a = atck.loc[atck['Age'] == elemento]
    lista_index2.append(int(a['index']))

print(lista_index2)
print(len(lista_index2))
```

```python
# Transformar cada string da lista em uma nova lista com 2 elementos:
# Cada elemento significa a idade individual de cada vítima de ataque
padrão = '[0-9][0-9]'
lista_ind2 = list()

for i in lista_mult2:
    a = re.findall(padrão, i)
    lista_ind2.append(a)

print(lista_ind2)
print(len(lista_ind2))
```

```python
def info_mult(linha):
    '''Criar uma lista com todas as informações do index em que tem mais de uma informação
    no campo AGE'''
    
    infos = list()
    global atck
    infos.append(atck.iloc[linha, 0])
    infos.append(atck.iloc[linha, 1])
    infos.append(atck.iloc[linha, 2])
    infos.append(atck.iloc[linha, 3])
    infos.append(atck.iloc[linha, 4])
    infos.append(atck.iloc[linha, 5])
    infos.append(atck.iloc[linha, 6])
    infos.append(atck.iloc[linha, 7])
    infos.append(atck.iloc[linha, 8])
    infos.append(atck.iloc[linha, 9])    
    
    return infos
```
